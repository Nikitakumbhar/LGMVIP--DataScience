# Importing Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Reading The Input Image

```python
img = cv2.imread(r"C:\Users\Hp\Desktop\flower-pots.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img)
plt.show()
```

# Converting Original Image to Grey Scale

```python

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# Inverting the Gray Image

```python
#Inversion of the image
inv_img = 255 - gray_img
plt.imshow(cv2.cvtColor(inv_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# Blurring the inverted gray image by Gausian Function

```python
#Blur the image 
blurred = cv2.GaussianBlur(inv_img, (21,21), 0)

#Invert the blurred image
inv_blur = 255 - blurred
final_img = cv2.divide(gray_img, inv_blur, scale=256.0)
plt.imshow(cv2.cvtColor(final_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
